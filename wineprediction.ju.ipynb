# Train and test the given dataset

```python
from sklearn.model_selection import train_test_split
x=df2.drop(['quality','Better_quality'],axis=1)
y=df2['Better_quality']
```

```python
x.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test.shape
```

Now we normalize the dataset to achieve stability and accurate prediction

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
X_train.shape
```

```python
X_train
```

Various model to predict accuracy of a given dataset

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
sv = SVC()
sv.fit(X_train,y_train)
sv.score(X_test,y_test)
```

```python
rf = RandomForestClassifier(n_estimators=10)
rf.fit(X_train,y_train)
rf.score(X_test,y_test)
```

```python
y_pred1=lr.predict(X_test)
y_pred2=sv.predict(X_test)
y_pred3=rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred1)
cm
```

```python
cm = confusion_matrix(y_test,y_pred2)
cm
```

```python
cm = confusion_matrix(y_test,y_pred3)
cm
```
